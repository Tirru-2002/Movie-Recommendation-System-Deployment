```python
headers = {"Accept-Language":"en-Us,en;q=0.5"}
for page in pages:
    conn = http.client.HTTPSConnection("www.imdb.com")
    conn.request("GET", "/list/ls503325184/?st_dt=&mode=detail&page="+str(page)+"&sort=list_order,asc")
    res = conn.getresponse()
    page = res.read()
    soup = BeautifulSoup(page, "html.parser")
    movie_data = soup.findAll("div",attrs = {"class":"lister-item mode-detail"})
    time.sleep(randint(2,8))

    for store in movie_data:
        name = store.h3.a.text
        movie_name.append(name)

        year_of_release = store.h3.find("span",class_ = "lister-item-year text-muted unbold").text.replace("(","").replace("II ","").replace("I ","").replace(")","")
        movie_year.append(year_of_release)

        try:
         runtime = store.p.find("span",class_="runtime").text.replace(",","")
        except:
         runtime=""

        Movie_Runtime.append(runtime)


        try:
         genre=store.p.find("span",class_="genre").text.replace("\n","")
        except:
          genre=""
        Genre.append(genre)

        try:
          age_rated = store.p.find("span",class_ = "certificate").text.replace("\n","")
        except:
          age_rated = ""
        Age_rated.append(age_rated)


        p_tags=store.find_all('p')
        try:
         desc=p_tags[1].text.replace("\n","")
        except:
         desc=""
        description.append(desc)

        directors=[]
        stars=[]
        director_links = p_tags[2].find_all('a', href=True)
        encounter=False

        for i, link in enumerate(director_links):
            text = link.get_text(strip=True)
            if i < len(director_links) - 1 and encounter is False:
             directors.append(text)
            #  print(text)
             if 'Stars' in director_links[i + 1].previous_sibling:
               encounter=True
            else:
             stars.append(text)

        Directors.append(directors)
        Stars.append(stars)

        # print("Directors:", directors)
        # print("Stars:", stars)

        rate = store.find("span",class_ = "ipl-rating-star__rating").text.replace("\n","")
        movie_rating.append(rate)

        meta = store.find("span",class_ = "metascore").text if store.find("span",class_ = "metascore") else ""
        movie_metascore.append(meta)

        value = store.find_all("span",attrs = {"name":"nv"})

        vote = value[0].text
        movie_votes.append(vote)

        gross = value[1].text.replace("$","").replace("M","")  if len(value) >1 else ""
        Gross_income.append(gross)

    conn.close()
```

```python
movie_list = pd.DataFrame({"Movie Name": movie_name,"Year of Release":movie_year,"Duration":Movie_Runtime,"Genre":Genre,"Certificate":Age_rated,"Movie Rating":movie_rating,"Metascore":movie_metascore,"Description":description,"Director":Directors,"Stars":Stars,"Votes":movie_votes,"Gross":Gross_income})
```
